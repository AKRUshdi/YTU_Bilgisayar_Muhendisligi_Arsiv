```python
# 1 Tokenizer yükleme

from transformers import BertTokenizer, AutoTokenizer

tokenizer = BertTokenizer.from_pretrained('dbmdz/bert-base-turkish-cased')

print(tokenizer)
```

```python
# 2. Metni Tokenlere Ayırma

text = "Merhaba, bu bir test cümlesidir."
tokens = tokenizer.tokenize(text)
print(tokens)

# 3. Tokenleri ID'lere Dönüştürme

token_ids = tokenizer.convert_tokens_to_ids(tokens)
print(token_ids)

# 4. ID'leri Tokenlere Dönüştürme

tokens_back = tokenizer.convert_ids_to_tokens(token_ids)
print(tokens_back)
```

```python
# 5. Metni Doğrudan Kodlama

encoded = tokenizer.encode(text, add_special_tokens=True)
print(encoded)

# 6. Encode Plus Kullanımı

encoded_plus = tokenizer.encode_plus(text, add_special_tokens=True, return_attention_mask=True)
print(encoded_plus)
```

```python
# 7. Batch Encode Plus ile Toplu Kodlama

texts = ["Merhaba dünya", "Bu bir test cümlesidir."]
batch_encoded = tokenizer.batch_encode_plus(texts, padding='max_length', return_attention_mask=True)
print(batch_encoded)
```

```python
# 8. Token ID'lerini Metne Çevirme

decoded_text = tokenizer.decode(encoded)
print(decoded_text)
```

```python
from datasets import disable_caching
disable_caching()
from datasets import load_dataset
import tensorflow as tf
from transformers import GPT2Config, TFAutoModelForCausalLM, GPT2TokenizerFast

tokenizer = GPT2TokenizerFast.from_pretrained("ytu-ce-cosmos/turkish-gpt2")
tokenizer.pad_token = tokenizer.eos_token

paths = [r"/content/drive/MyDrive/wikipedia_raw_0.csv"]


def remove_none(example):
    return example["Text"] is not None and example["Text"] != ''

block_size = 1024

raw_datasets = load_dataset('csv', data_files=paths)
raw_datasets['train'] = raw_datasets['train'].select(range(100))
raw_datasets = raw_datasets.filter(remove_none)

def tokenize_function(examples):
    return tokenizer(examples["Text"])

tokenized_datasets = raw_datasets.map(
    tokenize_function, batched=True, remove_columns=raw_datasets["train"].column_names
)

def split_into_subtokens(batch):
    max_length = 1024
    simplified_examples = {"input_ids": [], "attention_mask": [], "labels": []}

    for example in batch["input_ids"]:
        truncated_example = example[:max_length]

        padding_length = max_length - len(truncated_example)
        padded_example = truncated_example + [tokenizer.pad_token_id] * padding_length

        attention_mask = [1 if token_id != tokenizer.pad_token_id else 0 for token_id in padded_example]

        simplified_examples["input_ids"].append(padded_example)
        simplified_examples["attention_mask"].append(attention_mask)

    simplified_examples['labels'] = simplified_examples['input_ids'].copy()

    return simplified_examples

tokenized_datasets2 = tokenized_datasets.map(split_into_subtokens, batched=True, remove_columns=tokenized_datasets["train"].column_names)

config = GPT2Config.from_pretrained("ytu-ce-cosmos/turkish-gpt2")
config.bos_token_id = 0
config.eos_token_id = 0
config.pad_token_id = 0,
config.vocab_size = tokenizer.vocab_size
config.n_embd = 384
config.n_head = 6
config.layer = 6

model = TFAutoModelForCausalLM.from_config(config)
optimizer = tf.keras.optimizers.Adam(learning_rate=5e-5)
model.compile(optimizer=optimizer)

tf_train_dataset = model.prepare_tf_dataset(
    tokenized_datasets2["train"],
    shuffle=False,
    batch_size=4,
    )

model.fit(
    tf_train_dataset,
    epochs=1,
    steps_per_epoch=25,
)

model.save_pretrained("my_gpt_model")
```
